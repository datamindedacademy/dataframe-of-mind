```python
import polars as pl
import holidays
import datetime as dt
```

```python
df_visitors = pl.read_parquet("../data/visitors.parquet")
```

```python
def is_weekend(timestamp: dt.datetime) -> bool:
    """
    Determine whether a zoo opening date is a weekend day or not.
    """

    return timestamp.weekday() in [5, 6]


def is_holiday(timestamp: dt.datetime) -> bool:
    """
    Determine whether a zoo opening date is a holiday or not.
    """
    return timestamp.date() in holidays.country_holidays("BE")
```

In this exercise, we want to figure out which non-weekend and non-holiday days attracted the most visitors. To do so, we'll write two user-defined functions:


```python
(
    df_visitors.with_columns(
        is_weekend=pl.col("timestamp").map_elements(
            is_weekend, return_dtype=pl.Boolean
        ),
        is_holiday=pl.col("timestamp").map_elements(
            is_holiday, return_dtype=pl.Boolean
        ),
    )
    .filter(~pl.col("is_weekend") & ~pl.col("is_holiday"))
    .sort("visitors", descending=True)
    .head(5)
)
```
