```python
import polars as pl
```

```python
pl.read_csv(
    source="../data/measurements.csv",
    has_header=False,
    separator="|",
    infer_schema=True,
    try_parse_dates=True,
)
```

```python
schema = {
    "id": pl.String,
    "name": pl.String,
    "timestamp": pl.Datetime,
    "blood_pressure": pl.Int32,
    "heart_rate": pl.Int32,
    "temperature": pl.Float32,
    "blood_glucose": pl.Int32,
}
pl.read_csv(
    source="../data/measurements.csv",
    has_header=False,
    separator="|",
    schema=schema,
)
```
